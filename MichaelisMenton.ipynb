```python
from pylab import plot, show, title, xlabel, ylabel, legend
from numpy import arange

dt = 0.01
E0 = 1
S0 = 1
ES0 = 0
P0 = 0

kminus = 6
kplus = 2
r = 4

def dE(E,S,ES):
    return -kplus*E*S + kminus*ES + r*ES

def dS(E,S,ES):
    return -kplus*E*S + kminus*ES

def dES(E,S,ES):
    return kplus*E*S - kminus*ES - r*ES

def v(ES):
    return r*ES

t = arange(0,10,dt)

E1 = E0
S1 = S0
ES1 = ES0
P1 = P0

E = []
S = []
ES = []
P = []

for ti in t:
    E.append(E1)
    S.append(S1)
    ES.append(ES1)
    P.append(P1)
    E1 += dt*dE(E0,S0,ES0)
    S1 += dt*dS(E0,S0,ES0)
    ES1 += dt*dES(E0,S0,ES0)
    P1 += dt*v(ES0)
    E0 = E1
    S0 = S1
    ES0 = ES1
    P0 = P1

plot(t,E,label="[E]")
plot(t,S,label="[S]")
plot(t,ES,label="[ES]")
plot(t,P,label="[P]")
legend(loc='best')
title("Enyme Reaction progression")
xlabel("reaction time (s)")
ylabel("Species Concentraions (M)")
show()
```
