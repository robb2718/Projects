```python
from numpy import arange, array, pi, sin, cos
import matplotlib.pyplot as plt

g = 9.81
l = 0.1
theta0 = 179*pi/180
omega0 = 0
t0 = 0
tmax = 5000
dt = 1

def f(r):
    theta = r[0]
    omega = r[1]
    ftheta = omega
    fomega = -(g/l)*sin(theta)
    return array([ftheta,fomega], float)

tpoints = arange(t0,tmax,dt)

thetapoints = []

r = array([theta0,omega0],float)

for t in tpoints:
    thetapoints.append(r[0])
    k1 = dt*f(r)
    k2 = dt*f(r + 0.5*k1)
    k3 = dt*f(r + 0.5*k2)
    k4 = dt*f(r + k3)
    r += (k1 + 2*k2 + 2*k3 + k4)/6

plt.plot(tpoints,thetapoints)
plt.grid()
plt.show()
```
